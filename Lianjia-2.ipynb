```python
from sklearn.linear_model import LinearRegression
data = pd.read_csv('G:\\Python\\Lianjia.csv', encoding = 'gbk')
data.TotalPrice = data.TotalPrice.dropna()
data.Area = data.Area.dropna()
y = data.TotalPrice.reshape(-1,1)
x = data.Area.reshape(-1,1)
lin_reg = LinearRegression()
lin_reg.fit(x,y)
lin_reg.intercept_, lin_reg.coef_
x_new = np.array([[0],[500]])
x_new_b = np.c_[np.ones((2,1)),x_new]
y_predict = x_new_b * lin_reg.coef_+lin_reg.intercept_

plt.plot(x,y,"b.")
plt.plot(x_new_b, y_predict, "r-",linewidth = 2, label="Prediction")
plt.xlabel("$x_1$",fontsize=18)
plt.ylabel("$y$",rotation=0,fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0,500,0,3100])
plt.show()
```
